# 9. [BONUS] Plot X, Y and Best Fit Line

**Write a function to calculate best fit.**

![alt text](https://i.imgur.com/B43vOV0.jpg)

[Least Squares Regression (Best Line Fit)]: https://www.mathsisfun.com/data/least-squares-regression.html "Least Squares Regression"

```
# Expected Output:
```

````python
"""CODE DELETED. DO NOT TOUCH THIS TEXT. RUNNING THIS CELL WITH SHIFT-ENTER WILL DELETE THE GRAPH"""
````

````python
# sample points
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    N = len(X)
    sum_X = np.sum(X)
    sum_Y = np.sum(Y)
    sum_XY = np.sum(np.multiply(X, Y))
    sum_X_squared = np.sum(np.square(X))

    m = (N * sum_XY - sum_X * sum_Y) / (N * sum_X_squared - sum_X**2)
    b = (sum_Y - m * sum_X) / N

    return m, b

# solution
m, b = best_fit(X, Y)

# plot points and fit line
# ANSWER
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, m * np.array(X) + b, color='red', label=f'Best Fit Line: y = {m:.2f}x + {b:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Best Fit Line')
plt.legend()
plt.show()
````

````python

np.random.seed(0)
x = stats.loggamma.rvs(5, size=500) + 5
# x follows a loggamma (i.e. non-bell-shaped) distribution - hence its probplot will not be linear at either end
# A Box-Cox transformation of x can achieve a more bell-shaped-looking distribution
# ANSWER
fig = plt.figure()
ax1 = fig.add_subplot(211)
np.random.seed(0)
x = stats.loggamma.rvs(5, size=500) + 5

stats.probplot(x, dist="norm", plot=ax1)
ax1.set_title('Probplot of Original Data')
x_boxcox, _ = stats.boxcox(x)
ax2 = fig.add_subplot(212)
stats.probplot(x_boxcox, dist="norm", plot=ax2)
ax2.set_title('Probplot after Box-Cox Transformation')
plt.tight_layout()
plt.show()
````

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m, se

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m, se =  np.mean(a), scipy.stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h

data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
        167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]
print(mean_confidence_interval(data))
````
